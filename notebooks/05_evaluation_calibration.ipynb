# 05. Evaluation & Calibration
Analyzing model performance and calibration.

```
import sys
import matplotlib.pyplot as plt
sys.path.append('..')
from src.data.load_data import load_raw_data
from src.data.split_data import split_train_test
from src.features.preprocessing import preprocess_data
from src.models.train_logistic import train_logistic_regression
from src.evaluation.calibration import plot_calibration_curve_func
```

```
df = load_raw_data('../data/raw/credit_default_uci.xls')
X_train, X_test, y_train, y_test = split_train_test(df)
X_train_proc, X_test_proc, _ = preprocess_data(X_train, X_test)
```

```
lr = train_logistic_regression(X_train_proc, y_train)
y_prob = lr.predict_proba(X_test_proc)[:, 1]
```

```
plot_calibration_curve_func(y_test, y_prob, 'Logistic Regression')
plt.show()
```
